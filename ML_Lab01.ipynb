```python
import matplotlib.pyplot as plt

# Create random arrays from exponential, uniform, and normal distributions
exp_data = np.random.exponential(scale=1.0, size=100000)
uni_data = np.random.uniform(low=0.0, high=1.0, size=100000)
norm_data = np.random.normal(loc=0.0, scale=1.0, size=100000)

# Plot histograms
plt.hist(exp_data, density=True, bins=100, histtype="step", color="blue", label="Exponential")
plt.hist(uni_data, density=True, bins=100, histtype="step", color="green", label="Uniform")
plt.hist(norm_data, density=True, bins=100, histtype="step", color="red", label="Normal")
plt.legend(loc="upper right")
plt.title("Random Distributions")
plt.xlabel("Value")
plt.ylabel("Density")
plt.show()
```

```python
from mpl_toolkits.mplot3d import Axes3D

# Create data
x = np.linspace(-5, 5, 100)
y = np.linspace(-5, 5, 100)
x, y = np.meshgrid(x, y)
z = x**2 + y**2

# Plotting
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(x, y, z, cmap='viridis')

ax.set_title("3D Plot of Z = X^2 + Y^2")
ax.set_xlabel("X")
ax.set_ylabel("Y")
ax.set_zlabel("Z")

plt.show()
```

```python
import seaborn as sns
import pandas as pd

# Load the dataset
# Assume the seaborn tutorial dataset is available as a Pandas DataFrame
# data = pd.read_csv('path_to_dataset.csv')
# For demonstration, I'll use a sample DataFrame

data = pd.DataFrame({
    'HP': np.random.rand(100),
    'Attack': np.random.rand(100),
    'Defense': np.random.rand(100),
    'Sp. Atk': np.random.rand(100),
    'Sp. Def': np.random.rand(100),
    'Speed': np.random.rand(100)
})

# Calculate Pearson and Spearman correlations
pearson_corr = data.corr(method='pearson')
spearman_corr = data.corr(method='spearman')

# Visualize using heatmaps
plt.figure(figsize=(12, 6))

plt.subplot(1, 2, 1)
sns.heatmap(pearson_corr, annot=True, cmap='coolwarm')
plt.title('Pearson Correlation')

plt.subplot(1, 2, 2)
sns.heatmap(spearman_corr, annot=True, cmap='coolwarm')
plt.title('Spearman Correlation')

plt.show()
```

```python
#Task 2
#i
df_path = "/content/Au_nanoparticle_dataset.csv"
data = pd.read_csv(df_path)

filtered_data = data[['N_total', 'N_bulk', 'N_surface', 'R_avg']]
print(filtered_data.head(20))
```

```python
#ii
mean_values = filtered_data.mean()
std_values = filtered_data.std()
quartiles = filtered_data.quantile([0.25, 0.5, 0.75])

print("Means:\n", mean_values)
print("Standard Deviations:\n", std_values)
print("Quartiles:\n", quartiles)
```

```python
#iii
filtered_data.hist(bins=30, layout=(1, 4), figsize=(20, 5))
plt.suptitle("Histograms of Features")
plt.show()
```

```python
#iv
g = sns.pairplot(filtered_data)
g.map_diag(sns.histplot, kde=True)
g.map_lower(sns.kdeplot)
plt.show()
```

```python
#v
sns.pairplot(filtered_data)
plt.show()
```
